Here, I want to show the differences between assignment, copy and deepcopy

```python
import copy
```

```python
a=[1,2,[3]]
b=a
c=copy.copy(a)
d=copy.deepcopy(a)
print 'a:',a,'b:',b,'c:',c,'d:',d
```

```python
a.append(3)
print 'a:',a,'b:',b,'c:',c,'d:',d
```

```python
a[2].append(5)
print 'a:',a,'b:',b,'c:',c,'d:',d
```

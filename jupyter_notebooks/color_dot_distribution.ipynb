```python
%%writefile color_dot_distribution.py

import pylab as plt
import mpld3
import sys
mpld3.enable_notebook()
plt.rcParams['figure.figsize'] = 30, 20
plt.rcParams['font.size'] = 14.0
plt.rcParams['figure.subplot.top'] = 0.95
def check_germline_loh(input_filename,window=100000,minimum_size=1000000,allowed=5): #allowed_number_of_heter per window
    lines=open(input_filename,'r').readlines()
    length=len(lines)
    line_number=0
    list_loh=[]
    dic_loh={}

    window=int(window)
    minimum_size=int(minimum_size)
    minimum=minimum_size/window
    
    list_loh=[]
    fig,axs=plt.subplots(6,4)
    counter=0
    while line_number < 43:
        total_line=lines[line_number]
        freq_line=lines[line_number+1]
        chrom_num=line_number/2+1
        i=(chrom_num-1)/4
        j=(chrom_num-1) % 4
        chrom='chr%s' % (chrom_num)
        line_number+=2

        total=[int(count) for count in total_line.strip(" \r\n").split("\t")]
        
        record=freq_line.strip(" \r\n").split("\t")
        freq=[float(f) for f in record]
        axs[i,j].hist(freq,30, histtype='stepfilled', fc='lightblue', alpha=0.5)
        axs[i,j].set_title(chrom)
        counter+=1
    output_filename="histogram_%s.html" % (input_filename)
    mpld3.save_html(fig,output_filename)

if __name__=="__main__":
    #input_filename="loh_out_SJEWS010375_1M_color.txt"
    input_filename=sys.argv[1]
    check_germline_loh(input_filename)
 
```

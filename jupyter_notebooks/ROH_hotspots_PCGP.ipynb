```python
import sys
import pylab
import mpld3
import os
```

```python
mpld3.enable_notebook()
```

```python
ref_file="/home/zzhang1/karyotype/karyotype_hg19_modified.txt"
```

```python
os.chdir("/home/zzhang1/test_code/PCGP_hg19")
```

```python

pylab.rcParams['figure.figsize'] = 30, 20
pylab.rcParams['font.size'] = 14.0
pylab.rcParams['figure.subplot.top'] = 0.9

list_filenames=os.listdir(".")

filtered_filenames=[filename for filename in list_filenames if filename.startswith("summ") and filename.endswith("100K.txt")]

num_of_files=len(filtered_filenames)

chromosome_info={}

for i in range(22):
    chrom="chr"+str(i+1)
    chromosome_info[chrom]={}
chromosome_info['chrX']={}
chromosome_info['chrY']={}


ref=open(ref_file,'r')
lines=ref.readlines()
for line in lines:
    if line.startswith("#"):
        pass
    else:
        record=line.strip(" \r\n").split()
        if len(record)<=1:
            pass
        elif record[-1]=="acen":
            chrom="chr"+record[1]
            start=int(record[4])
            end=int(record[5])
            chromosome_info[chrom]['cent_start']=start
            chromosome_info[chrom]['cent_end']=end
        elif record[-1]=="yellow":
            chrom="chr"+record[1]
            start=int(record[4])
            end=int(record[5])
            chromosome_info[chrom]['gap_start']=start
            chromosome_info[chrom]['gap_end']=end
        elif record[-1] in ["vdgrey","black"]:
            chrom="chr"+record[2]
            length=int(record[5])
            chromosome_info[chrom]['length']=length
        else:
            pass


dic_out={}

for key,value in chromosome_info.iteritems():
    length=value['length']/100000 #1000000=window
    dic_out[key]=[0]  #[heter/(heter+homo), # of heter, # of homo SNPS]
    for i in range(length):
        dic_out[key].append(0)

for summ_file in filtered_filenames: 
    #Sample  Chr     Length(kb)      Start(kb)       End(kb)
    #loh_germline_SJTALL005_color_selection_100K.txt chr6    7800    44700   52500
    file1=open(summ_file,'r')
    summary=file1.readlines()[1:]
    if len(summary)<=0:
        pass
    else:
        for summ in summary: 
            record=summ.strip(" \r\n").split("\t")
            chrom=record[1]
            spread=int(record[2])/100 #100=window/1000
            start=int(record[3])/100  #100=window/1000
            for i in range(spread-1):
                dic_out[chrom][start+1+i]+=1
    
    
list_chrom=["chr"+str(i+1) for i in range(22)]

fig, ax = pylab.subplots(6,4)

counter=0
for chrom in list_chrom:
    i=counter/4
    j=counter%4

    print "plotting figure No.",str(counter)
    x_values=[v+1 for v in range(len(dic_out[chrom]))]
    y_values=[item for item in dic_out[chrom]] #percentage of heterozygosites
    
    scatter=ax[i,j].scatter(x_values,y_values,linewidths=0,alpha=0.5) #s=sizes,c=colors
    subtitle="%s" % (chrom)
    ax[i,j].set_title(subtitle)
    
    counter+=1

```

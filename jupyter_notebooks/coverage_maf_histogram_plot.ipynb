```python
import pylab as plt
import numpy as np
from numpy import median,mean
from scipy.stats import gaussian_kde
import mpld3
#mpld3.enable_notebook()
def coverage_plot(filename):
    input1=open(filename,'r')
    header_line=input1.readline()
    header=header_line.strip(" \r\n").split("\t")
    index_ref=header.index("reference_normal_count")
    index_alt=header.index("alternative_normal_count")
    index_name=header.index("Name")
    line=input1.readline()
    coverages=[]
    mafs=[]
    labels=[]
    count=0
    while line and count<500:
    #while line:
        record=line.strip(" \r\n").split("\t")
        count_ref=int(record[index_ref])
        count_alt=int(record[index_alt])
        name=record[index_name]
        coverage=count_ref+count_alt
        #if coverage < 100:
        coverages.append(coverage)
        labels.append(name)
        if coverage>0:
            maf=float(count_ref)/coverage
        else:
            maf=0
        mafs.append(maf)
        count+=1
        line=input1.readline()
    xy=np.vstack([coverages,mafs])
    z=gaussian_kde(xy)(xy)
    fig,ax=plt.subplots()
    scatter=ax.scatter(coverages,mafs,c=z,s=5,edgecolor='')
    #plt.savefig("SJACT065_coverage_maf.png")
    tooltip=mpld3.plugins.PointLabelTooltip(scatter,labels=labels)
    mpld3.plugins.connect(fig,tooltip)
    #mpld3.save_html(fig,"SJACT065_coverage_maf.html")
    #plt.hist(coverages,300)
    #print median(coverages)
    #return coverages
filename="/nfs_exports/genomes/1/PCGP/BucketIntermediate/SJACT/SnpDetect/SJACT065_bam_high_20.out"
coverages=coverage_plot(filename)

```

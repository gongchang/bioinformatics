```python
%matplotlib inline
import pylab as plt
import mpld3
mpld3.enable_notebook()
```

```python
import os
hg18=os.listdir("/nfs_exports/genomes/1/Homo_sapiens/1000Genomes/lowcoverage_with_window")
print hg18[19]
direct=" "
```

```python
import numpy as np
dic_sample={}
#Sample	Chr	Length(kb)	Start(kb)	End(kb)
#loh_germline_SJTALL005_color_selection_1M.txt	chr11	16000	98000	114000
for filename in hg18:
    if filename.startswith("summary_loh_germline") and filename.endswith("100K.txt"):
        #print filename
        fullname=direct+filename
        #print fullname
        sample=filename.split("_")[3]
        lines=open(fullname,'r').readlines()
        if len(lines)==1:
            dic_sample[sample]=0
        else:
            for line in lines[1:]:
                record=line.strip(" \r\n").split("\t")
                length=int(record[2])
                if sample not in dic_sample:
                    dic_sample[sample]=length
                else:
                    dic_sample[sample]+=length
#print dic_sample['SJMB013']
fig,ax=plt.subplots()
list1=dic_sample.values()
print len(list1)
ax.hist(dic_sample.values(),bins=np.logspace(-1, 20.0, 500)) #,bins=np.logspace(-1, 20.0, 500)
ax.set_title("total length of LOH with minimum length of 1MB (1000genomes hg18)")
```

```python
mpld3.save_html(fig,"1000genomes_total_length_hg18_179_samples.html")
```

```python
hg19=hg18
direct2=direct
```

```python
import numpy as np
plt.rcParams['figure.figsize'] = 30, 20
plt.rcParams['font.size'] = 14.0
plt.rcParams['figure.subplot.top'] = 0.9
dic_sample2={}
#Sample	Chr	Length(kb)	Start(kb)	End(kb)
#loh_germline_SJTALL005_color_selection_1M.txt	chr11	16000	98000	114000
for filename in hg19:
    if filename.startswith("summary_loh_germline") and filename.endswith("100K.txt"):
        #print filename
        fullname=direct2+filename
        #print fullname
        sample=filename.split("_")[3]
        lines=open(fullname,'r').readlines()
        if len(lines)==1:
            dic_sample2[sample]=[0,0,[]]
        else:
            for line in lines[1:]:
                record=line.strip(" \r\n").split("\t")
                length=int(record[2])
                chrom=record[1]
                if sample not in dic_sample2:
                    dic_sample2[sample]=[length,1,[chrom]]   # length, count of different chroms, [chroms]
                else:
                    if chrom not in dic_sample2[sample][2]:
                        dic_sample2[sample][0]+=length
                        dic_sample2[sample][1]+=1
                        dic_sample2[sample][2].append(chrom)
                    else:
                        dic_sample2[sample][0]+=length
                        

fig2,(ax1, ax2) = plt.subplots(2, 1)
total_length=[item[0] for item in dic_sample2.values()]
ax1.hist(total_length,bins=np.logspace(-1, 20.0, 500))
ax1.grid(color='yellow',linestyle='solid')
ax1.set_title("total length of LOH with minimum length of 1MB (PCGP hg19)")

labels=[]
x_values=[]
y_values=[]
colors=[]
positive_1M=['SJACT065', 'SJACT069', 'SJACT071', 'SJBALL020013', 'SJBALL020984', 'SJERG002', 'SJEWS010365', 'SJEWS010367', 'SJEWS010375', 'SJHGG033', 'SJHGG040', 'SJHGG059A', 'SJHGG102', 'SJHYPER043', 'SJHYPER084', 'SJHYPER146', 'SJHYPER150', 'SJMB032', 'SJOS001101', 'SJOS001109']
for key,value in dic_sample2.iteritems():
    labels.append(key)
    x_values.append(value[0])
    y_values.append(value[1])
    if key in positive_1M:
        colors.append("red")
    else:
        colors.append("blue")

scatter=ax2.scatter(x_values,y_values,c=colors,linewidths=0,alpha=0.5) #s=sizes
#ax2.set_xscale('log')
ax2.set_xlabel('total length of LOH')
ax2.set_ylabel('total numbers of chr with LOH >1MB')
tooltip=mpld3.plugins.PointLabelTooltip(scatter,labels=labels)
mpld3.plugins.connect(fig2,tooltip)
```
